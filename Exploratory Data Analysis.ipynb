# Exploratory Data Analysis

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1.1 Import Data

```python
train = pd.read_csv("train.csv")
val = pd.read_csv("val.csv")
explore = pd.concat([train, val], axis=0)

explore.head()
```

# 2.1 Univariate Analysis

Now I will inspect the distribution of each feature

```python
explore.hist(figsize=(15, 23));
```

It seems that some classes are rather rare. Also Ba, K, and Fe are often zero. This could be a useful insight in feature engineering.

# 3.1 Multivariate Analysis

Now I will plot every feature against every other feature to see if there are interesting interactions.

```python

# logic for not plotting same relationship twice
columns_used = []
for col1 in explore.columns:
    for col2 in explore.columns:
        same = (col1 == col2)
        plotted_before = (col2 in columns_used)
        columns_used.append(col2)
        
        if ((not same) & (not plotted_before)):
            
            # plotting
            plt.figure(figsize=(15, 5))
            plt.scatter(explore[col1], explore[col2])
            plt.title(f"{col1} VS {col2}")
            plt.show()
```

It seems there is a lack of visible relationships going on between the features.

# 3.2 Correlation

Now I will measure the pearson correlation for each feature and the target. I will plot the results.

```python
corr = explore.corr()["Type"]

plt.figure(figsize=(15, 5))
plt.bar(explore.columns, corr)
plt.title("Correlation of features to the target")
plt.show()

plt.figure(figsize=(15, 5))
plt.bar(explore.columns, abs(corr))
plt.show("Absolute value of correlation")
plt.show()
```
